# El teorema de Bayes para una media normal con una prior discreta

## Con una única observación

Cojamos una única observación proveniente de la función de densidad $f(y|\mu)$ que suponemos
normal con varianza conocida $\sigma^2$. La desviación estándar $\sigma$ es la raíz cuadrada de la varianza y para
la media, tenemos m valores posibles: $\mu_1,…,\mu_m$. A partir de estos valores se elije una distribución
prior que represente nuestra creencia antes de tomar la observación. En el caso de que no se
tuviese la información prior, daríamos la misma probabilidad de éxito a todos los valores.

La verosimilitud otorga pesos a todos los valores posibles según su probabilidad de ocurrencia. Para
ello fijamos el valor observado y variamos la variable entre todos los valores posibles. La distribución
posterior es proporcional al producto de la prior y la verosimilitud:

$$g(\mu|y) = \frac{prior * likelihood}{\sum prior * likelihood}$$

## La verosimilitud de una única observación

La distribución condicional de $y|\mu$ es normal con media $\mu$, varianza conocida $\sigma^2$ y tiene la siguiente
función de densidad:

$$f(y|\mu) = \frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2\sigma^2}(y-\mu)^2}$$

La forma de la verosimilitud viene definida por:

$$f(y|\mu) \propto e^{-\frac{1}{2\sigma^2}(y-\mu)^2}$$

## Con una muestra aleatoria

Normalmente, en vez de una única observación, se tienen $y_1, y_2,...,y_n$ observaciones provenientes de una muestra aleatoria donde cada distribución $y_j|\mu$ es normal con media $\mu$ y varianza conocida $\sigma^2$. La distribución posterior es proporcional al producto de la prior y la verosimilitud y como las observaciones de la muestra son independientes entre sí, la verosimilitud conjunta viene dada por el producto de la verosimilitud de cada observación:

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) = 𝑓(𝑦_1|𝜇) ∗ 𝑓(𝑦_2|𝜇) ∗. . .∗ 𝑓(𝑦_𝑛|𝜇).$$

Por lo tanto, en este caso el teorema de Bayes con una previa discreta se expresa:

$$𝑔(𝜇|𝑦_1, . . . , 𝑦_𝑛) ∝ 𝑔(𝜇) ∗ 𝑓(𝑦_1|𝜇) ∗ 𝑓(𝑦_2|𝜇) ∗. . .∗ 𝑓(𝑦_𝑛|𝜇)$$.

### Probabilidades posteriores

Las probabilidades posteriores se pueden buscar una a una de forma secuencial o todas a la vez. En
el primer caso, la posterior de una observación se convierte en la prior de la siguiente observación
pero este método puede ser muy costoso a medida que se incrementa el tamaño muestral. En el
segundo caso, como cada observación es normal, cada una de ellas tiene una verosimilitud normal y
la verosimilitud conjunta viene dada por:

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) ∝ e^{-\frac{1}{2\sigma^2}(y_1-\mu)^2} * e^{-\frac{1}{2\sigma^2}(y_2-\mu)^2} * ... * e^{-\frac{1}{2\sigma^2}(y_2 - \mu)^2},$$

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) ∝ e^{-\frac{1}{2\sigma^2}[\displaystyle \sum_{i=1}^n(y_i-\mu)^2]}$$

Centrándonos en los términos de dentro del paréntesis cuadrados:

$$[(𝑦_1 − 𝜇)^2 + (𝑦_2 − 𝜇)^2 + ⋯ + (𝑦_𝑛 − 𝜇)^2] = 𝑦_1^2 − 2𝑦_1𝜇 + 𝜇^2 + ⋯ + 𝑦_𝑛^2 − 2𝑦_𝑛𝜇 + 𝜇^2 =$$
$$= (𝑦_1^2 + ⋯ + 𝑦_𝑛^2) − 2𝜇(𝑦_1 + ⋯ + 𝑦_𝑛) + 𝑛𝜇^2 = 𝑛[\frac{(𝑦_1^2 + ⋯ + 𝑦_𝑛^2)}{n}-2𝜇\bar{y}+ 𝜇^2]$$

Entonces,

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) ∝ 𝑒^{-\frac{n}{2\sigma^2}[\displaystyle 𝜇^2−2𝜇\bar{y}+ \bar{y}^2−\bar{y}^2 + \frac{(𝑦_1^2 + ⋯ + 𝑦_𝑛^2)}{n}]}$$

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) ∝ 𝑒^{-\frac{n}{2\sigma^2}[\displaystyle 𝜇^2−2𝜇\bar{y}+ \bar{y}^2]}𝑒^{-\frac{n}{2\sigma^2}[\displaystyle \frac{(𝑦_1^2 + ⋯ + 𝑦_𝑛^2)}{n} −\bar{y}^2]} $$

La verosimilitud de una muestra normal aleatoria $𝑦_1, 𝑦_2, . . . , 𝑦_𝑛$ es proporcional a la verosimilitud de
la media muestral $\bar{y}$. Simplificando la parte que no depende del parámetro $𝜇$,

$$𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇) ∝ 𝑒^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}.$$

Se observa que esta verosimilitud tiene forma de una normal con media $𝜇$ y varianza $\frac{𝜎^2}{n}$.

## El teorema de Bayes para una media normal con una previa continua

En el apartado anterior los valores que podía tomar $𝜇$ estaban restringidos a un número
determinado de valores, pero es más realista pensar que $𝜇$ puede tomar cualquier valor, o al menos,
todos los que estén incluidos en un cierto intervalo. En estos casos, conviene utilizar una distribución
prior continua. Podemos escribir el teorema de Bayes de la siguiente manera:

$$𝑔(𝜇|𝑦_1, . . . , 𝑦_𝑛) ∝ 𝑔(𝜇) ∗ 𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇)$$

Cuando la prior era discreta, para conseguir la posterior teníamos que dividir el producto de la
verosimilitud y la prior por la suma de este producto para todos los valores de $𝜇$. Cuando la
distribución es continua, el procedimiento equivalente es hacer la integral de ese producto. Es decir:

$$𝑔(𝜇|𝑦_1, . . . , 𝑦_𝑛) = \frac{𝑔(𝜇) ∗ 𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇)}{\int 𝑔(𝜇) ∗ 𝑓(𝑦_1, . . . , 𝑦_𝑛|𝜇)𝑑𝜇}$$

Para una distribución normal, la verosimilitud de la muestra aleatoria es proporcional a la
verosimilitud de la media muestral $\bar{y}$. Entonces,

$$𝑔(𝜇|𝑦1, . . . , 𝑦𝑛) = \frac{g(\mu)*𝑒^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}}{\int g(\mu)*𝑒^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}d\mu}$$




### Densidad prior plana para 𝜇 (Previa de Jeffrey)

La prior plana considera que todos los valores que puede tomar 𝜇 son equiprobables, y entrega a
todos el mismo peso: 𝑔(𝜇) = 1. La prior plana no es realmente una distribución prior ya que la
integral del área no es 1, pero esto no es un problema. Además, la distribución a posteriori sí que
integrará 1, que es lo que nos interesa.


### Una observación normal

Sea 𝑦 una observación distribuida como una normal de media 𝜇 y varianza $𝜎^2$. Ignorando la constante de proporcionalidad, la verosimilitud viene dada por:

$$𝑓(𝑦|𝜇) = 𝑒^{\displaystyle -\frac{1}{2\sigma^2}(y-\mu)^2}.$$

Como la prior siempre es 1, la posterior es proporcional a la expresión anterior:

$$g(𝜇|y) = 𝑒^{\displaystyle -\frac{1}{2\sigma^2}(\mu - y)^2}.$$

A partir de esta ecuación, podemos concluir que la posterior sigue una distribución normal de media 𝑦 y varianza $𝜎^2$.

### Una muestra aleatoria normal

En uno de los apartados anteriores vimos que la verosimilitud de una muestra aleatoria proveniente
de una distribución normal, era proporcional a la verosimilitud de la media muestral $\bar{y}$. Se sabe que
$\bar{y}$ está distribuido de manera normal con media 𝜇 y varianza $𝜎^2/n$. Entonces, si ignoramos la constante,
la verosimilitud viene dada por:

$$𝑓(𝑦|𝜇) = 𝑒^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}.$$

Como la prior siempre es 1, la posterior es proporcional a la expresión anterior:

$$g(𝜇|y) = 𝑒^{\displaystyle -\frac{1}{2\sigma^2/n}(\mu - \bar{y})^2}.$$

A partir de esta ecuación, podemos concluir que la posterior sigue una distribución normal de media $\bar{y}$ y varianza
$𝜎^2/n$.

## Densidad prior normal para 𝜇

### Una observación normal

Sea una observación 𝑦 proveniente de una distribución normal con media 𝜇 y varianza conocida $𝜎^2$.
Supongamos que tenemos una prior normal con media 𝑚 y varianza $𝑠^2$. Si ignoramos la parte que
no contiene a 𝜇, las formas de la previa y de la verosimilitud vienen dadas por:

$$𝑔(𝜇) ∝ 𝑒^{\displaystyle -\frac{1}{2s^2}(𝜇−𝑚)^2},$$

$$𝑓(𝑦|𝜇) ∝ 𝑒^{\displaystyle -\frac{1}{2𝜎^2}(𝑦 −𝜇)^2}.$$

El producto de ambas ecuaciones se representa:

$$𝑔(𝜇) ∗ 𝑓(𝑦 |𝜇) ∝ e^{\displaystyle -\frac{1}{2}[\frac{(\mu-m)^2}{s^2}+\frac{(y-\mu)^2}{\sigma^2}]}$$

$$∝ e^{\displaystyle  -\frac{1}{2}[\frac{𝜎^2(𝜇^2−2𝜇𝑚+𝑚^2)+𝑠^2(𝑦^2−2𝑦𝜇+𝜇^2)}{𝑠^2𝜎^2}]}$$

$$∝ e^{\displaystyle  -\frac{1}{2}[\frac{(𝜎^2+𝑠^2)𝜇^2−2(𝜎^2𝑚+𝑠^2𝑦)𝜇+𝑚^2𝜎^2+𝑦^2𝑠^2)}{𝑠^2𝜎^2}]}$$

Si sacamos el término $\frac{(𝜎^2+𝑠^2)}{𝜎^2𝑠^2}$ como factor común, utilizamos las fórmulas notables y absorbemos
los términos que no dependen de 𝜇, tenemos que:

$$∝ e^{\displaystyle  -\frac{1}{2𝜎^2𝑠^2/(𝜎^2+𝑠^2)}[\mu^2 - 2\frac{(𝜎^2𝑚+𝑠^2𝑦)}{s^2+\sigma^2}𝜇+(\frac{𝜎^2𝑚+𝑠^2𝑦}{s^2+\sigma^2})^2]}$$

$$∝ e^{\displaystyle  -\frac{1}{2𝜎^2𝑠^2/(𝜎^2+𝑠^2)}[\mu -(\frac{𝜎^2𝑚+𝑠^2𝑦}{s^2+\sigma^2})]^2}$$

A partir de aquí, podemos concluir que la distribución posterior tiene la siguiente media y la
siguiente varianza:

$$m'=\frac{𝜎^2𝑚 + 𝑠^2𝑦}{𝜎^2 + 𝑠^2}$$

$$(s')^2 = \frac{𝜎^2𝑠^2}{𝜎^2 + 𝑠^2}$$

Es decir, hemos empezado con una previa $𝑁(𝑚, 𝑠^2)$ y acabamos con una posterior $𝑁(𝑚′, (𝑠′)^2)$.
Esto implica que al aplicar el teorema de Bayes, no será necesario hacer la integral ya que se trata de
dos distribuciones de la misma familia.

### Una muestra aleatoria normal

Sea una muestra aleatoria $𝑦_1, 𝑦_2, . . . , 𝑦_𝑛$ que viene de una distribución normal con media 𝜇 y
varianza conocida $𝜎^2$. Tenemos la siguiente distribución prior normal con media m y varianza $𝑠^2$:

$$𝑔(𝜇) ∝ 𝑒^{-\frac{1}{2s^2}(\mu-m)^2}$$

Usaremos la verosimilitud de la media muestral $\bar{y}$, la cual se distribuye como una normal con media
𝜇 y varianza conocida $𝜎^2/𝑛$.

Derivando los calculos como el caso anterior, la distribution posterior corresponde a una normal:

$$N(\mu_1=\frac{\tau_0\mu_0 + n\tau\bar{y}}{\tau_0 + n\tau},\tau_1= \tau_0+n\tau)$$

donde $\tau_0 = \frac{1}{s^2}$, $\tau = \frac{1}{\sigma^2}$, $\mu_0 = m$

# Conjugación: Distribuciones conjugadas

Definición: Supongamos que una distribución a prior pertenece a cierta clase o familia de distribuciones paramétricas D; entonces diremos que esa distribución es conjugada respecto a la verosimilitud si la distribución a posteriori también pertenece a la clase D.

| prior | verosimilitud | posteriori |
| :------- | :------: | -----: |
| Beta   | Binomial      | Beta  |
| Gamma   | Poisson    | Gamma |
| Normal   | Normal($\tau=1/\sigma^2$ conocida)    | Normal |


### (verosimilitud Binomial)

En un proceso experimental de Bernouilli, el número de éxitos X en n pruebas independientes de resultado
dicotómico A y no-A con $P(a)= \theta$ sigue una distribución binomial: $X \sim B(n,\theta)$

Por tanto dada una experiencia que arroja x éxitos la verosimilitud será:

$$f(x|\theta) = \binom{n}{x}\theta^x (1-\theta)^{n-x} ∝\theta^x (1-\theta)^{n-x}$$

ya que $\binom{n}{x}$ no depende ni del parámetro $\theta$ ni de los datos x, podemos incorporarlos en la constante de
normalización.

Nos planteamos cómo incorporar la información inicial sobre $g$ para por realizar inferencias.

1. Como el parámetro $\theta$ es una tasa, una proporción o una probabilidad tendremos que $\theta \in [0,1]$ y que se deberá cumplir que la densidad a prior $g(\theta)$:
$$\int_0^1 g(\theta)d\theta=1$$

una densidad beta cumple esto.
2. Debe ser conjugada a la d. binomial: si $g(\theta) \simBeta(\alpha_0,\beta_0)$ la densidad a posteriori también debe seguir una Beta:

$$g(\theta|x) ∝ g(\theta)f(x|\theta) \rightarrow Beta(\alpha_1,\beta_1)$$

$$g(\theta|x) ∝ \theta^{(x+\alpha_0)-1}(1-\theta)^{(n-x+\beta_0)-1}$$



```python
# veamos un simple ejemplo del lanzamiento de una moneda.
# La idea principal de la inferencia bayesiana es que la noción de probabilidad cambia mientras más datos tengamos.


import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas

np.random.seed(2131982) # para poder replicar el random

# Graficos embebidos.
%matplotlib inline 


import matplotlib.pyplot as plt # importando matplotlib
import seaborn as sns # importando seaborn

# parametros esteticos de seaborn
sns.set_palette("deep", desat=.6)
#sns.set_context(rc={"figure.figsize": (15, 15)})
plt.figure(figsize=(10, 8))

dist = stats.beta
n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)


for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$p$, probabilidad de cara") \
        if k in [0, len(n_trials) - 1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="lanzamientos observados %d,\n %d caras" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)

plt.suptitle("Actualizacion Bayesiana de probabilidades posterios",y=1.02,fontsize=14)

plt.tight_layout()
```

```python

n_trials = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
data = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]
theta_real = 0.35

beta_params = [(1, 1), (0.5, 0.5), (20, 20)]
dist = stats.beta
x = np.linspace(0, 1, 2000)
plt.figure(figsize=(10, 8))

for idx, N in enumerate(n_trials):
    if idx == 0:
        plt.subplot(4, 3, 2)
        plt.xlabel('θ')
    else:
        plt.subplot(4, 3, idx+3)
        plt.xticks([])
    y = data[idx]
    for (a_prior, b_prior) in beta_params:
        p_theta_given_y = dist.pdf(x, a_prior + y, b_prior + N - y)
        plt.fill_between(x, 0, p_theta_given_y, alpha=0.7)

    plt.plot(theta_real, 0, ms=9, marker='o', mec='w', mfc='k')
    plt.plot(0, 0, label=f'{N:4d} experimentos\n{y:4d} caras', alpha=0)
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()
    plt.yticks([])
plt.tight_layout()
```

# Analizando los resultados

La primer figura del panel muestra los a prior, nuestra estimación de $\theta$ dado que no hemos realizado ningún experimento. Las sucesivas nueve figuras muestran las distribuciones a posteriori y se indica la cantidad de experimentos y de caras obtenidas. Además se puede ver una linea negra vertical en 0.35, la cual representa el valor verdadero de $\theta$. Por supuesto que en problemas reales este valor es desconocido.

Este ejemplo es realmente ilustrativo en varios aspectos.

1. El resultado de un análisis Bayesiano NO es un solo valor, si no una distribución (a posteriori) de los valores plausibles de los parámetros (dado los datos y el modelo).
2. La dispersión o ancho de las curvas es una medida de la incertidumbre sobre los valores.
3. El valor más probable viene dado por la moda de la distribución (la cumbre de la distribución).
4. Aún cuando $\frac{2}{1} = \frac{8}{4}$ son numéricamente iguales tenemos menor incertidumbre en un resultado cuando el número de experimentos es mayor.
5. Dada una cantidad suficiente de datos los resultados tienden a converger sin importar el a prior usado.
6. La rapidez con la que los resultados convergen varía. En este ejemplo las curvas azul y anaranjada parecen converger con tan solo 8 experimentos, pero se necesitan más de 50 experimentos para que las tres curvas se muestren similares. Aún con 150 experimentos se observan ligeras diferencias.
7. Partiendo de los a prior uniforme (azul) o sesgado (anaranjado) y habiendo realizado un solo experimento y observado una sola cara, lo más razonable es pensar que estamos frente a una moneda con dos caras!
8. La situación cambia drásticamente al ver por primera vez una moneda caer ceca. Ahora lo más probable (dado cualquiera de los tres a prioris) es inferir que $\theta=0.5$. Los valores de $\theta$ exactamente 0 o 1 se vuelven imposibles.
9. El a prior no sesgado (verde) es más informativo que los otros dos (la distribución esta más concentrada), por ello se requiere de un número mas grande de experimentos para "moverlo".
10. El a prior uniforme (azul) es lo que se conoce como no informativo. El resultado de un análisis Bayesiano usando un a priori no-informativos en general coinciden con los resultados de análisis frecuentistas (en este caso el valor esperado de $\theta = \frac{y}{N}$).

Resumiendo el a posteriori
El resultado de un análisis Bayesiano es siempre una distribución de probabilidad. En el caso de la moneda esto es evidente, y en el caso del diagnostico es menos claro ya que la distribución es discreta y solo puede tomar dos valores.

A la hora de comunicar los resultados de un análisis Bayesiano, lo más informativo es reportar la distribución completa, aunque esto no siempre es posible o deseable, por ejemplo el a posteriori de una distribución multidimensional es imposible de dibujar en papel. En general, se suele recurrir a distintas medidas que resumen el a prior, por ejemplo reportando la media de la distribución a posteriori. Algo un poco más informativo es reportar además un intervalo de credibilidad. Existen varios criterios para definir intervalos de credibilidad, el que usaremos en este curso (y que también es ampliamente usado en la literatura) es lo que se conoce como intervalo de más alta densidad y nos referiremos a él por su sigla en ingles, HPD (Highest Posterior Density interval). Un HPD es el intervalo, más corto, que contiene una porción fija de la densidad de probabilidad, generalmente el 95% (aunque otros valores como 90% o 50% son comunes). Cualquier punto dentro de este intervalo tiene mayor densidad que cualquier punto fuera del intervalo. Para una distribución unimodal, el HPD 95 es simplemente el intervalo entre los percentiles 2,5 y 97,5.

# Posteriori

La distribución de probabilidad resultante que resume tanto los datos anteriores como los anteriores es la posterior.

## INTERVALO DE ALTA DENSIDAD POSTERIOR (HPD)

Una herramienta muy útil para resumir la propagación de la densidad posterior. Se define como el intervalo más corto que contiene una porción dada de densidad de probabilidad.

Hay varias formas de calcular Computacional posterior que se pueden clasificar ampliamente como

1. Métodos no markovianos
2. Métodos markovianos

### MÉTODOS NO MARKOVIANOS

COMPUTACIÓN DE LA RED: es un enfoque de fuerza bruta utilizado principalmente cuando no podemos calcular analíticamente la parte posterior. Para un modelo de parámetro único, el enfoque de cuadrícula es el siguiente:

1. Defina un intervalo razonable para el parámetro (la prior debería darle una pista).
2. Coloque una cuadrícula de puntos (generalmente equidistantes) en ese intervalo.
3. Para cada punto en la cuadrícula, multiplicamos la verosimilitud y la prior.
4. Normalice los valores calculados (divida el resultado en cada punto por la suma de todos los puntos).

El enfoque de computación en cuadrícula no escala bien para datos de alta dimensión.

1. Use la siguiente función para explorar diversas combinaciones de priors y likelihoods. Enuncie las conclusiones que considere más relevantes.

```python
def a_posteriori_grilla(grilla=10, a=1, b=1, caras=6, tiradas=9):
    grid = np.linspace(0, 1, grilla)
    prior = stats.beta(a, b).pdf(grid)
    likelihood = stats.binom.pmf(caras, tiradas, grid)
    posterior = likelihood * prior
    posterior /= posterior.sum()
    _, ax = plt.subplots(1, 3, sharex=True, figsize=(16, 4))
    ax[0].set_title('caras = {}\ntiradas = {}'.format(caras, tiradas))
    for i, (e, e_n) in enumerate(zip([prior, likelihood, posterior], ['a priori', 'likelihood', 'a posteriori'])):
        ax[i].set_yticks([])
        ax[i].plot(grid, e, 'o-', label=e_n)
        ax[i].legend(fontsize=14)


interact(a_posteriori_grilla, grilla=ipyw.IntSlider(min=2, max=100, step=1, value=15), a=ipyw.FloatSlider(min=1, max=7, step=1, value=1), b=ipyw.FloatSlider(
    min=1, max=7, step=1, value=1), caras=ipyw.IntSlider(min=0, max=20, step=1, value=6), tiradas=ipyw.IntSlider(min=0, max=20, step=1, value=9));
```
