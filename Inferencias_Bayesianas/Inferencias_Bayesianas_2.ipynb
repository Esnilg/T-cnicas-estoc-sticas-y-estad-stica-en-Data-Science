# El teorema de Bayes para una media normal con una prior discreta

## Con una √∫nica observaci√≥n

Cojamos una √∫nica observaci√≥n proveniente de la funci√≥n de densidad $f(y|\mu)$ que suponemos
normal con varianza conocida $\sigma^2$. La desviaci√≥n est√°ndar $\sigma$ es la ra√≠z cuadrada de la varianza y para
la media, tenemos m valores posibles: $\mu_1,‚Ä¶,\mu_m$. A partir de estos valores se elije una distribuci√≥n
prior que represente nuestra creencia antes de tomar la observaci√≥n. En el caso de que no se
tuviese la informaci√≥n prior, dar√≠amos la misma probabilidad de √©xito a todos los valores.

La verosimilitud otorga pesos a todos los valores posibles seg√∫n su probabilidad de ocurrencia. Para
ello fijamos el valor observado y variamos la variable entre todos los valores posibles. La distribuci√≥n
posterior es proporcional al producto de la prior y la verosimilitud:

$$g(\mu|y) = \frac{prior * likelihood}{\sum prior * likelihood}$$

## La verosimilitud de una √∫nica observaci√≥n

La distribuci√≥n condicional de $y|\mu$ es normal con media $\mu$, varianza conocida $\sigma^2$ y tiene la siguiente
funci√≥n de densidad:

$$f(y|\mu) = \frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2\sigma^2}(y-\mu)^2}$$

La forma de la verosimilitud viene definida por:

$$f(y|\mu) \propto e^{-\frac{1}{2\sigma^2}(y-\mu)^2}$$

## Con una muestra aleatoria

Normalmente, en vez de una √∫nica observaci√≥n, se tienen $y_1, y_2,...,y_n$ observaciones provenientes de una muestra aleatoria donde cada distribuci√≥n $y_j|\mu$ es normal con media $\mu$ y varianza conocida $\sigma^2$. La distribuci√≥n posterior es proporcional al producto de la prior y la verosimilitud y como las observaciones de la muestra son independientes entre s√≠, la verosimilitud conjunta viene dada por el producto de la verosimilitud de cada observaci√≥n:

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) = ùëì(ùë¶_1|ùúá) ‚àó ùëì(ùë¶_2|ùúá) ‚àó. . .‚àó ùëì(ùë¶_ùëõ|ùúá).$$

Por lo tanto, en este caso el teorema de Bayes con una previa discreta se expresa:

$$ùëî(ùúá|ùë¶_1, . . . , ùë¶_ùëõ) ‚àù ùëî(ùúá) ‚àó ùëì(ùë¶_1|ùúá) ‚àó ùëì(ùë¶_2|ùúá) ‚àó. . .‚àó ùëì(ùë¶_ùëõ|ùúá)$$.

### Probabilidades posteriores

Las probabilidades posteriores se pueden buscar una a una de forma secuencial o todas a la vez. En
el primer caso, la posterior de una observaci√≥n se convierte en la prior de la siguiente observaci√≥n
pero este m√©todo puede ser muy costoso a medida que se incrementa el tama√±o muestral. En el
segundo caso, como cada observaci√≥n es normal, cada una de ellas tiene una verosimilitud normal y
la verosimilitud conjunta viene dada por:

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) ‚àù e^{-\frac{1}{2\sigma^2}(y_1-\mu)^2} * e^{-\frac{1}{2\sigma^2}(y_2-\mu)^2} * ... * e^{-\frac{1}{2\sigma^2}(y_2 - \mu)^2},$$

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) ‚àù e^{-\frac{1}{2\sigma^2}[\displaystyle \sum_{i=1}^n(y_i-\mu)^2]}$$

Centr√°ndonos en los t√©rminos de dentro del par√©ntesis cuadrados:

$$[(ùë¶_1 ‚àí ùúá)^2 + (ùë¶_2 ‚àí ùúá)^2 + ‚ãØ + (ùë¶_ùëõ ‚àí ùúá)^2] = ùë¶_1^2 ‚àí 2ùë¶_1ùúá + ùúá^2 + ‚ãØ + ùë¶_ùëõ^2 ‚àí 2ùë¶_ùëõùúá + ùúá^2 =$$
$$= (ùë¶_1^2 + ‚ãØ + ùë¶_ùëõ^2) ‚àí 2ùúá(ùë¶_1 + ‚ãØ + ùë¶_ùëõ) + ùëõùúá^2 = ùëõ[\frac{(ùë¶_1^2 + ‚ãØ + ùë¶_ùëõ^2)}{n}-2ùúá\bar{y}+ ùúá^2]$$

Entonces,

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) ‚àù ùëí^{-\frac{n}{2\sigma^2}[\displaystyle ùúá^2‚àí2ùúá\bar{y}+ \bar{y}^2‚àí\bar{y}^2 + \frac{(ùë¶_1^2 + ‚ãØ + ùë¶_ùëõ^2)}{n}]}$$

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) ‚àù ùëí^{-\frac{n}{2\sigma^2}[\displaystyle ùúá^2‚àí2ùúá\bar{y}+ \bar{y}^2]}ùëí^{-\frac{n}{2\sigma^2}[\displaystyle \frac{(ùë¶_1^2 + ‚ãØ + ùë¶_ùëõ^2)}{n} ‚àí\bar{y}^2]} $$

La verosimilitud de una muestra normal aleatoria $ùë¶_1, ùë¶_2, . . . , ùë¶_ùëõ$ es proporcional a la verosimilitud de
la media muestral $\bar{y}$. Simplificando la parte que no depende del par√°metro $ùúá$,

$$ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá) ‚àù ùëí^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}.$$

Se observa que esta verosimilitud tiene forma de una normal con media $ùúá$ y varianza $\frac{ùúé^2}{n}$.

## El teorema de Bayes para una media normal con una previa continua

En el apartado anterior los valores que pod√≠a tomar $ùúá$ estaban restringidos a un n√∫mero
determinado de valores, pero es m√°s realista pensar que $ùúá$ puede tomar cualquier valor, o al menos,
todos los que est√©n incluidos en un cierto intervalo. En estos casos, conviene utilizar una distribuci√≥n
prior continua. Podemos escribir el teorema de Bayes de la siguiente manera:

$$ùëî(ùúá|ùë¶_1, . . . , ùë¶_ùëõ) ‚àù ùëî(ùúá) ‚àó ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá)$$

Cuando la prior era discreta, para conseguir la posterior ten√≠amos que dividir el producto de la
verosimilitud y la prior por la suma de este producto para todos los valores de $ùúá$. Cuando la
distribuci√≥n es continua, el procedimiento equivalente es hacer la integral de ese producto. Es decir:

$$ùëî(ùúá|ùë¶_1, . . . , ùë¶_ùëõ) = \frac{ùëî(ùúá) ‚àó ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá)}{\int ùëî(ùúá) ‚àó ùëì(ùë¶_1, . . . , ùë¶_ùëõ|ùúá)ùëëùúá}$$

Para una distribuci√≥n normal, la verosimilitud de la muestra aleatoria es proporcional a la
verosimilitud de la media muestral $\bar{y}$. Entonces,

$$ùëî(ùúá|ùë¶1, . . . , ùë¶ùëõ) = \frac{g(\mu)*ùëí^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}}{\int g(\mu)*ùëí^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}d\mu}$$




### Densidad prior plana para ùúá (Previa de Jeffrey)

La prior plana considera que todos los valores que puede tomar ùúá son equiprobables, y entrega a
todos el mismo peso: ùëî(ùúá) = 1. La prior plana no es realmente una distribuci√≥n prior ya que la
integral del √°rea no es 1, pero esto no es un problema. Adem√°s, la distribuci√≥n a posteriori s√≠ que
integrar√° 1, que es lo que nos interesa.


### Una observaci√≥n normal

Sea ùë¶ una observaci√≥n distribuida como una normal de media ùúá y varianza $ùúé^2$. Ignorando la constante de proporcionalidad, la verosimilitud viene dada por:

$$ùëì(ùë¶|ùúá) = ùëí^{\displaystyle -\frac{1}{2\sigma^2}(y-\mu)^2}.$$

Como la prior siempre es 1, la posterior es proporcional a la expresi√≥n anterior:

$$g(ùúá|y) = ùëí^{\displaystyle -\frac{1}{2\sigma^2}(\mu - y)^2}.$$

A partir de esta ecuaci√≥n, podemos concluir que la posterior sigue una distribuci√≥n normal de media ùë¶ y varianza $ùúé^2$.

### Una muestra aleatoria normal

En uno de los apartados anteriores vimos que la verosimilitud de una muestra aleatoria proveniente
de una distribuci√≥n normal, era proporcional a la verosimilitud de la media muestral $\bar{y}$. Se sabe que
$\bar{y}$ est√° distribuido de manera normal con media ùúá y varianza $ùúé^2/n$. Entonces, si ignoramos la constante,
la verosimilitud viene dada por:

$$ùëì(ùë¶|ùúá) = ùëí^{\displaystyle -\frac{1}{2\sigma^2/n}(\bar{y}-\mu)^2}.$$

Como la prior siempre es 1, la posterior es proporcional a la expresi√≥n anterior:

$$g(ùúá|y) = ùëí^{\displaystyle -\frac{1}{2\sigma^2/n}(\mu - \bar{y})^2}.$$

A partir de esta ecuaci√≥n, podemos concluir que la posterior sigue una distribuci√≥n normal de media $\bar{y}$ y varianza
$ùúé^2/n$.

## Densidad prior normal para ùúá

### Una observaci√≥n normal

Sea una observaci√≥n ùë¶ proveniente de una distribuci√≥n normal con media ùúá y varianza conocida $ùúé^2$.
Supongamos que tenemos una prior normal con media ùëö y varianza $ùë†^2$. Si ignoramos la parte que
no contiene a ùúá, las formas de la previa y de la verosimilitud vienen dadas por:

$$ùëî(ùúá) ‚àù ùëí^{\displaystyle -\frac{1}{2s^2}(ùúá‚àíùëö)^2},$$

$$ùëì(ùë¶|ùúá) ‚àù ùëí^{\displaystyle -\frac{1}{2ùúé^2}(ùë¶ ‚àíùúá)^2}.$$

El producto de ambas ecuaciones se representa:

$$ùëî(ùúá) ‚àó ùëì(ùë¶ |ùúá) ‚àù e^{\displaystyle -\frac{1}{2}[\frac{(\mu-m)^2}{s^2}+\frac{(y-\mu)^2}{\sigma^2}]}$$

$$‚àù e^{\displaystyle  -\frac{1}{2}[\frac{ùúé^2(ùúá^2‚àí2ùúáùëö+ùëö^2)+ùë†^2(ùë¶^2‚àí2ùë¶ùúá+ùúá^2)}{ùë†^2ùúé^2}]}$$

$$‚àù e^{\displaystyle  -\frac{1}{2}[\frac{(ùúé^2+ùë†^2)ùúá^2‚àí2(ùúé^2ùëö+ùë†^2ùë¶)ùúá+ùëö^2ùúé^2+ùë¶^2ùë†^2)}{ùë†^2ùúé^2}]}$$

Si sacamos el t√©rmino $\frac{(ùúé^2+ùë†^2)}{ùúé^2ùë†^2}$ como factor com√∫n, utilizamos las f√≥rmulas notables y absorbemos
los t√©rminos que no dependen de ùúá, tenemos que:

$$‚àù e^{\displaystyle  -\frac{1}{2ùúé^2ùë†^2/(ùúé^2+ùë†^2)}[\mu^2 - 2\frac{(ùúé^2ùëö+ùë†^2ùë¶)}{s^2+\sigma^2}ùúá+(\frac{ùúé^2ùëö+ùë†^2ùë¶}{s^2+\sigma^2})^2]}$$

$$‚àù e^{\displaystyle  -\frac{1}{2ùúé^2ùë†^2/(ùúé^2+ùë†^2)}[\mu -(\frac{ùúé^2ùëö+ùë†^2ùë¶}{s^2+\sigma^2})]^2}$$

A partir de aqu√≠, podemos concluir que la distribuci√≥n posterior tiene la siguiente media y la
siguiente varianza:

$$m'=\frac{ùúé^2ùëö + ùë†^2ùë¶}{ùúé^2 + ùë†^2}$$

$$(s')^2 = \frac{ùúé^2ùë†^2}{ùúé^2 + ùë†^2}$$

Es decir, hemos empezado con una previa $ùëÅ(ùëö, ùë†^2)$ y acabamos con una posterior $ùëÅ(ùëö‚Ä≤, (ùë†‚Ä≤)^2)$.
Esto implica que al aplicar el teorema de Bayes, no ser√° necesario hacer la integral ya que se trata de
dos distribuciones de la misma familia.

### Una muestra aleatoria normal

Sea una muestra aleatoria $ùë¶_1, ùë¶_2, . . . , ùë¶_ùëõ$ que viene de una distribuci√≥n normal con media ùúá y
varianza conocida $ùúé^2$. Tenemos la siguiente distribuci√≥n prior normal con media m y varianza $ùë†^2$:

$$ùëî(ùúá) ‚àù ùëí^{-\frac{1}{2s^2}(\mu-m)^2}$$

Usaremos la verosimilitud de la media muestral $\bar{y}$, la cual se distribuye como una normal con media
ùúá y varianza conocida $ùúé^2/ùëõ$.

Derivando los calculos como el caso anterior, la distribution posterior corresponde a una normal:

$$N(\mu_1=\frac{\tau_0\mu_0 + n\tau\bar{y}}{\tau_0 + n\tau},\tau_1= \tau_0+n\tau)$$

donde $\tau_0 = \frac{1}{s^2}$, $\tau = \frac{1}{\sigma^2}$, $\mu_0 = m$

# Conjugaci√≥n: Distribuciones conjugadas

Definici√≥n: Supongamos que una distribuci√≥n a prior pertenece a cierta clase o familia de distribuciones param√©tricas D; entonces diremos que esa distribuci√≥n es conjugada respecto a la verosimilitud si la distribuci√≥n a posteriori tambi√©n pertenece a la clase D.

| prior | verosimilitud | posteriori |
| :------- | :------: | -----: |
| Beta   | Binomial      | Beta  |
| Gamma   | Poisson    | Gamma |
| Normal   | Normal($\tau=1/\sigma^2$ conocida)    | Normal |


### (verosimilitud Binomial)

En un proceso experimental de Bernouilli, el n√∫mero de √©xitos X en n pruebas independientes de resultado
dicot√≥mico A y no-A con $P(a)= \theta$ sigue una distribuci√≥n binomial: $X \sim B(n,\theta)$

Por tanto dada una experiencia que arroja x √©xitos la verosimilitud ser√°:

$$f(x|\theta) = \binom{n}{x}\theta^x (1-\theta)^{n-x} ‚àù\theta^x (1-\theta)^{n-x}$$

ya que $\binom{n}{x}$ no depende ni del par√°metro $\theta$ ni de los datos x, podemos incorporarlos en la constante de
normalizaci√≥n.

Nos planteamos c√≥mo incorporar la informaci√≥n inicial sobre $g$ para por realizar inferencias.

1. Como el par√°metro $\theta$ es una tasa, una proporci√≥n o una probabilidad tendremos que $\theta \in [0,1]$ y que se deber√° cumplir que la densidad a prior $g(\theta)$:
$$\int_0^1 g(\theta)d\theta=1$$

una densidad beta cumple esto.
2. Debe ser conjugada a la d. binomial: si $g(\theta) \simÔÅæBeta(\alpha_0,\beta_0)$ la densidad a posteriori tambi√©n debe seguir una Beta:

$$g(\theta|x) ‚àù g(\theta)f(x|\theta) \rightarrow Beta(\alpha_1,\beta_1)$$

$$g(\theta|x) ‚àù \theta^{(x+\alpha_0)-1}(1-\theta)^{(n-x+\beta_0)-1}$$



```python
# veamos un simple ejemplo del lanzamiento de una moneda.
# La idea principal de la inferencia bayesiana es que la noci√≥n de probabilidad cambia mientras m√°s datos tengamos.


import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas

np.random.seed(2131982) # para poder replicar el random

# Graficos embebidos.
%matplotlib inline 


import matplotlib.pyplot as plt # importando matplotlib
import seaborn as sns # importando seaborn

# parametros esteticos de seaborn
sns.set_palette("deep", desat=.6)
#sns.set_context(rc={"figure.figsize": (15, 15)})
plt.figure(figsize=(10, 8))

dist = stats.beta
n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)


for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$p$, probabilidad de cara") \
        if k in [0, len(n_trials) - 1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="lanzamientos observados %d,\n %d caras" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)

plt.suptitle("Actualizacion Bayesiana de probabilidades posterios",y=1.02,fontsize=14)

plt.tight_layout()
```

```python

n_trials = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
data = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]
theta_real = 0.35

beta_params = [(1, 1), (0.5, 0.5), (20, 20)]
dist = stats.beta
x = np.linspace(0, 1, 2000)
plt.figure(figsize=(10, 8))

for idx, N in enumerate(n_trials):
    if idx == 0:
        plt.subplot(4, 3, 2)
        plt.xlabel('Œ∏')
    else:
        plt.subplot(4, 3, idx+3)
        plt.xticks([])
    y = data[idx]
    for (a_prior, b_prior) in beta_params:
        p_theta_given_y = dist.pdf(x, a_prior + y, b_prior + N - y)
        plt.fill_between(x, 0, p_theta_given_y, alpha=0.7)

    plt.plot(theta_real, 0, ms=9, marker='o', mec='w', mfc='k')
    plt.plot(0, 0, label=f'{N:4d} experimentos\n{y:4d} caras', alpha=0)
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()
    plt.yticks([])
plt.tight_layout()
```

Adem√°s se puede ver una linea negra vertical en 0.35, la cual representa el valor verdadero de $\theta$. Por supuesto que en problemas reales este valor es desconocido.

1. El resultado de un an√°lisis Bayesiano NO es un solo valor, si no una distribuci√≥n (a posteriori) de los valores plausibles de los par√°metros (dado los datos y el modelo).
2. La dispersi√≥n o ancho de las curvas es una medida de la incertidumbre sobre los valores.
3. El valor m√°s probable viene dado por la moda de la distribuci√≥n (la cumbre de la distribuci√≥n).
4. A√∫n cuando $\frac{2}{1} = \frac{8}{4}$ son num√©ricamente iguales tenemos menor incertidumbre en un resultado cuando el n√∫mero de experimentos es mayor.
5. Dada una cantidad suficiente de datos los resultados tienden a converger sin importar el a prior usado.
6. La rapidez con la que los resultados convergen var√≠a. En este ejemplo las curvas azul y anaranjada parecen converger con tan solo 8 experimentos, pero se necesitan m√°s de 50 experimentos para que las tres curvas se muestren similares. A√∫n con 150 experimentos se observan ligeras diferencias.
7. Partiendo de los a prior uniforme (azul) o sesgado (anaranjado) y habiendo realizado un solo experimento y observado una sola cara, lo m√°s razonable es pensar que estamos frente a una moneda con dos caras!
8. La situaci√≥n cambia dr√°sticamente al ver por primera vez una moneda caer ceca. Ahora lo m√°s probable (dado cualquiera de los tres a prioris) es inferir que $\theta=0.5$. Los valores de $\theta$ exactamente 0 o 1 se vuelven imposibles.
9. El a prior no sesgado (verde) es m√°s informativo que los otros dos (la distribuci√≥n esta m√°s concentrada), por ello se requiere de un n√∫mero mas grande de experimentos para "moverlo".
10. El a prior uniforme (azul) es lo que se conoce como no informativo. El resultado de un an√°lisis Bayesiano usando un a priori no-informativos en general coinciden con los resultados de an√°lisis frecuentistas (en este caso el valor esperado de $\theta = \frac{y}{N}$).

Resumiendo el a posteriori
El resultado de un an√°lisis Bayesiano es siempre una distribuci√≥n de probabilidad. En el caso de la moneda esto es evidente, y en el caso del diagnostico es menos claro ya que la distribuci√≥n es discreta y solo puede tomar dos valores.

A la hora de comunicar los resultados de un an√°lisis Bayesiano, lo m√°s informativo es reportar la distribuci√≥n completa, aunque esto no siempre es posible o deseable, por ejemplo el a posteriori de una distribuci√≥n multidimensional es imposible de dibujar en papel. En general, se suele recurrir a distintas medidas que resumen el a prior, por ejemplo reportando la media de la distribuci√≥n a posteriori. Algo un poco m√°s informativo es reportar adem√°s un intervalo de credibilidad. Existen varios criterios para definir intervalos de credibilidad, el que usaremos en este curso (y que tambi√©n es ampliamente usado en la literatura) es lo que se conoce como intervalo de m√°s alta densidad y nos referiremos a √©l por su sigla en ingles, HPD (Highest Posterior Density interval). Un HPD es el intervalo, m√°s corto, que contiene una porci√≥n fija de la densidad de probabilidad, generalmente el 95% (aunque otros valores como 90% o 50% son comunes). Cualquier punto dentro de este intervalo tiene mayor densidad que cualquier punto fuera del intervalo. Para una distribuci√≥n unimodal, el HPD 95 es simplemente el intervalo entre los percentiles 2,5 y 97,5.

Por ejemplo plot_posterior puede ser usado para generar un gr√°fico con la media y HPD. Una herramienta muy √∫til para resumir la propagaci√≥n de la densidad posterior. Se define como el intervalo m√°s corto que contiene una porci√≥n dada de densidad de probabilidad.

### M√âTODOS NO MARKOVIANOS

COMPUTACI√ìN DE LA RED: es un enfoque de fuerza bruta utilizado principalmente cuando no podemos calcular anal√≠ticamente la parte posterior. Para un modelo de par√°metro √∫nico, el enfoque de cuadr√≠cula es el siguiente:

1. Defina un intervalo razonable para el par√°metro (la prior deber√≠a darle una pista).
2. Coloque una cuadr√≠cula de puntos (generalmente equidistantes) en ese intervalo.
3. Para cada punto en la cuadr√≠cula, multiplicamos la verosimilitud y la prior.
4. Normalice los valores calculados (divida el resultado en cada punto por la suma de todos los puntos).

El enfoque de computaci√≥n en cuadr√≠cula no escala bien para datos de alta dimensi√≥n.

1. Use la siguiente funci√≥n para explorar diversas combinaciones de priors y likelihoods. Enuncie las conclusiones que considere m√°s relevantes.

```python
def a_posteriori_grilla(grilla=10, a=1, b=1, caras=6, tiradas=9):
    grid = np.linspace(0, 1, grilla)
    prior = stats.beta(a, b).pdf(grid)
    likelihood = stats.binom.pmf(caras, tiradas, grid)
    posterior = likelihood * prior
    posterior /= posterior.sum()
    _, ax = plt.subplots(1, 3, sharex=True, figsize=(16, 4))
    ax[0].set_title('caras = {}\ntiradas = {}'.format(caras, tiradas))
    for i, (e, e_n) in enumerate(zip([prior, likelihood, posterior], ['a priori', 'likelihood', 'a posteriori'])):
        ax[i].set_yticks([])
        ax[i].plot(grid, e, 'o-', label=e_n)
        ax[i].legend(fontsize=14)


interact(a_posteriori_grilla, grilla=ipyw.IntSlider(min=2, max=100, step=1, value=15), a=ipyw.FloatSlider(min=1, max=7, step=1, value=1), b=ipyw.FloatSlider(
    min=1, max=7, step=1, value=1), caras=ipyw.IntSlider(min=0, max=20, step=1, value=6), tiradas=ipyw.IntSlider(min=0, max=20, step=1, value=9));
```
